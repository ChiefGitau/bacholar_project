```python
def get_age_cat(num):
    if num < 10:
     return '0-10'
    elif num < 20:
     return '10-20'
    elif num < 30:
     return '20-30'
    elif num < 40:
     return '30-40'
    elif num < 50:
     return '40-50'
    elif num < 60:
     return '50-60'
    elif num < 70:
     return '60-70'
    elif num < 80:
     return '70-80'
    elif num < 90:
     return '80-90'
    else:
     return '90+'
```

```python
import ast

def make_age_gender(df_age_mid, df_age):
    for index, row in df_age_mid.iterrows():
        # female = row['female']
        # male = row['male']
        #
        # row['age_group_a']
        if row['number'] == 0:
            continue
        else:
            x = row['json_response']
            value_x = ast.literal_eval(x)
            new_value = value_x['data']

            for el in value_x['data']['detections']:
                gender_value = el['gender']
                age_value = el['age']
                aprox = int((age_value['minimum'] + age_value['maximum']) / 2)

                temp_group = get_age_cat(aprox)


                if gender_value['gender'] == 'Female':
                    df_age.loc[df_age['age'] == temp_group, 'female'] =  df_age['female'].loc[df_age['age'] == temp_group].item() + 1
                else:
                    df_age.loc[df_age['age'] == temp_group, 'male'] =  df_age['male'].loc[df_age['age'] == temp_group].item() + 1

    return df_age


```

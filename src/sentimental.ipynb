```python
import sys
mean_file = 'ilsvrc_2012_mean.npy'
import caffe
import numpy as np

model_loc = 'twitter_finetuned_test4_iter_180_conv.caffemodel'
model_deploy ='sentiment_maps/sentiment_fully_conv_deploy.prototxt'


model =caffe.Net(model_deploy,
                           model_loc,  caffe.TEST)

transformer = caffe.io.Transformer({'data': model.blobs['data'].data.shape})
transformer.set_mean('data', np.load(mean_file).mean(1).mean(1))
transformer.set_transpose('data', (2,0,1))
transformer.set_channel_swap('data', (2,1,0))
transformer.set_raw_scale('data', 255.0)


```

```python
def Average(lst):
        return sum(lst) / len(lst)
```

```python


im_caffe = caffe.io.load_image('3.2.jpg')
im_caffe_2 = caffe.io.load_image('3.3.jpg')



out = model.forward_all(data=np.asarray([transformer.preprocess('data', im_caffe)]))
out_2 = model.forward_all(data=np.asarray([transformer.preprocess('data', im_caffe_2)]))

# Save the 2x8x8 prediction
image_name = "fuuuuck"
np.save('test_2', out['prob'][0])

el_test = out['prob'][0][1]
el = out_2['prob'][0][1]

total_val = 0
print(len(el))
for i, ls in enumerate(el):
    test_av = 0
    for j, el_sub in enumerate(ls):
        test_av += abs(el_sub -  el_test[i][j])
    print('first kernal value is -> ' + str(test_av))
    total_val += (test_av/len(ls))
    print('difference in row value is ' + str(total_val))
print(total_val/len(el))
```
